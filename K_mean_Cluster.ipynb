```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import numpy as np
import sklearn
```

```python
url = 'https://github.com/prasertcbs/basic-dataset/raw/master/iris.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.columns
```

```python
sns.pairplot(df , vars =['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
             hue = 'species',
             plot_kws = {'alpha' : .5})
```

```python
sns.scatterplot(data =df ,x='petal_length',y='petal_width',hue='species')
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters = 3)
model
```

```python
x = df[['petal_length','petal_width']]
model.fit(x)
```

```python
center = model.cluster_centers_
```

```python
sns.scatterplot(data = df, x='petal_length',y='petal_width',hue ='species', alpha = .3)
plt.scatter(center[:,0],center[:,1],color ='red')
```

```python
model.labels_
```

```python
df['cluster'] = model.labels_
df[:5]
```

```python
pd.crosstab(df['species'],df['cluster'])
```

```python
fig , ax = plt.subplots(1,2 ,figsize = (10,5), sharey = True , sharex = True)
g1 = sns.scatterplot(data = df, x='petal_length',y='petal_width',hue ='species', 
                     alpha = .3 , ax=ax[0])
g1.scatter(center[:,0],center[:,1],color ='red')

g2 = sns.scatterplot(data = df, x='petal_length',y='petal_width',hue ='cluster',
                     palette = ['blue','orange','green'] ,
                     alpha = .3 , ax=ax[1])
g2.scatter(center[:,0],center[:,1],color ='red')


```
